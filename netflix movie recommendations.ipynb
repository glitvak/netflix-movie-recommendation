```python
# load data
data = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')

# print the head to see what we're working with

data.head()
```

```python
# replaces null values with unknown

data = data.replace(np.nan, 'Unknown', regex = True)

data.head()
```

```python
# select columns to use for reccomendations

col = ['type', 'director', 'cast', 'country', 'listed_in', 'title']
```

```python
# function for using the data to get reccommended. This is a string with the columns we are going to use to base reccomendations off

def important_features(x):
    features = []
    for i in range(0, x.shape[0]):
        features.append(x['cast'][i] + ' ' + x['director'][i] + ' ' + x['listed_in'][i] + ' ' + x['title'][i] + ' ' + x['country'][i] + ' ' + x['type'][i])
        
    return features
```

```python
data['features'] = important_features(data)

data.head()
```

```python
# tokenize the collection of text we made with t he features column
from sklearn.feature_extraction.text import CountVectorizer
cm = CountVectorizer().fit_transform(data['features'])
```

```python
# use cosine similarity to figure out simialrities between each movie row
from sklearn.metrics.pairwise import cosine_similarity

cs = cosine_similarity(cm)
# closer to 1 == really similar
# closer to 0 == not similar at all
print(cs)
```

```python
# to test, lets get a random movie/show title

title = 'Naruto Shippuden : Blood Prison'
movie_id = data[data.title == title].index.values.astype(int)[0]
print(movie_id)
```

```python
# create a list for similarity score and sort it
scores = sorted(list(enumerate(cs[movie_id])), key = lambda x:x[1], reverse = True)
scores = scores[1:]
```

```python
#show the top 5 reccomended movies

for i in range(0, 5):
    title1 = data.iloc[scores[i][0]]['title']
    print(title1)
```
